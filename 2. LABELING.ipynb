### Menghapus data yang ada di label

```python
import os

def delete_same_row(sub_df, full_df):
    sub_df = sub_df.set_index('raw')
    full_df = full_df.set_index('raw')
    list_index_df1 = list(sub_df.index)
    list_index_df2 = list(full_df.index)
    same_kalimat = []
    for kalimat in list_index_df1:
        if kalimat in list_index_df2:
            same_kalimat.append(kalimat)
    
    print(same_kalimat)
    final_df = full_df.copy()
    final_df = final_df.drop(same_kalimat)
    return full_df


csv_data_label = 'C:\\Users\\user\\PENELITIAN SAINS DATA\\BARU\\data\\manualLabelV1\\FinalLabel.csv' #@param {type:"string"}
directory = 'C:\\Users\\user\\PENELITIAN SAINS DATA\\BARU\\data\\processedV4' #@param {type:"string"}
out_directory = 'C:\\Users\\user\\PENELITIAN SAINS DATA\\BARU\\data\\nonLabelV1' #@param {type:"string"}

if directory[len(directory)-1] != '/':
    directory = directory + '/'

if out_directory[len(out_directory)-1] != '/':
    out_directory = out_directory + '/'

files = os.listdir(directory)
df_label = pd.read_csv(csv_data_label)

for file in files:
    if file[len(file)-4:] == '.csv':
        print(file)
        df = pd.read_csv(directory+file)
        df_temp = delete_same_row(df_label, df)
        df_temp.to_csv(out_directory+file)
```
