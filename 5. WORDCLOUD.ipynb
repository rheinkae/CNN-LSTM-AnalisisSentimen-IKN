```python
from wordcloud import WordCloud
wordcloud = WordCloud(width=1600, height=800, max_font_size=200, background_color='white')

import matplotlib.pyplot as plt
import pandas as pd
```

```python
df_data = pd.read_csv('C:\\Users\\user\\PENELITIAN SAINS DATA\\data\\processedV4\\data_mentah.csv')
```

### WordCloud 

```python
word = {}
for i in range(len(df_data)):
  temp_text = df_data.loc[i]['processed']
  temp_split = temp_text.split(' ')
  temp_set = list(set(temp_split))
  for j in range(len(temp_set)):
    try:
      word[temp_set[j]] += 1
    except:
      word[temp_set[j]] = 1

word_sorted = dict(sorted(word.items(), key=lambda x: x[1], reverse=True))
word_sorted
     
```

```python
wordcloud.generate_from_frequencies(word_sorted)
plt.figure(figsize=(12,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
df_total = pd.DataFrame(list(word_sorted.items()),columns = ['word','count'])
df_total.to_csv("wordcount.csv")
```
